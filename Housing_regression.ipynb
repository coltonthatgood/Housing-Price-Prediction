```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('C:\\Users\\USER\\.cache\\kagglehub\\datasets\\yasserh\\housing-prices-dataset\\versions\\1\\Housing.csv')
```

```python
print(df)
```

```python
df = df.dropna()
df['price'] = df['price'].astype(float)
```

```python
print(df)
```

```python
x = df[['area','bedrooms','bathrooms','stories','parking']]
print(x)
```

```python
y = df['price']
print(y)
```

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
predict_price = regr.predict([[6000,3,2,2,2]]) 
print(predict_price)
```
